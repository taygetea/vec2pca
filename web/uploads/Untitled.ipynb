```python
%matplotlib notebook
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt

model = Word2Vec.load("../../100features_100min_words_10context")
keys = list(model.vocab.keys())
df = pd.DataFrame(model[keys], index=keys)

pca = PCA(n_components=50)
pc = pca.fit_transform(df)
component_names = ["PC"+str(x) for x in range(1,len(pc[0])+1)]
df2 = pd.DataFrame(pc, index=df.index, columns=component_names)

css = """
$pink: #e91e63;
$cubic: cubic-bezier(.64,.09,.08,1);
* {
  box-sizing: border-box;
  font-family: Roboto, sans-serif;
}
body > div {
  margin-bottom: 30px;
  background: white;
  padding: 10px;
  border-radius: 2px;
  box-shadow: 2px 2px 4px 0px rgba(0,0,0,0.2);
  &:last-of-type {
    margin-bottom: 100px;
  }
}
p {
  display: inline;
  margin-right: 50px;
  font-size: 200%;
}
div.mpld3-tooltip {
  background: #616161;
  color: white;
  font-size: 200%;
  opacity: 0.9;
  padding-left: 8dp;
  padding-right: 8dp;
}
"""

df3 = df2.sample(n=200)
x=df3['PC1']
y=df3['PC2']

fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'))

scatter = ax.scatter(x, y, s=180, c="#90CAF9", )
plt.xlabel('PC1', size=20)
plt.ylabel('PC2', size=20)
ax.grid(color='white', linestyle='solid')
ax.set_title("Words by component vector", size=20)

labels = list(df3.index)
tooltip = mpld3.plugins.PointHTMLTooltip(scatter, labels=labels, 
                                         css=css)
mpld3.plugins.connect(fig, tooltip)
mpld3.display()
#html = mpld3.fig_to_html(fig)

#with open("lw_comp_23.html", "w") as f:
#    f.write(html)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mpld3
css = """
$pink: #e91e63;
$cubic: cubic-bezier(.64,.09,.08,1);
* {
  box-sizing: border-box;
  font-family: Roboto, sans-serif;
}
body > div {
  margin-bottom: 30px;
  background: white;
  padding: 10px;
  border-radius: 2px;
  box-shadow: 2px 2px 4px 0px rgba(0,0,0,0.2);
  &:last-of-type {
    margin-bottom: 100px;
  }
}
p {
  display: inline;
  margin-right: 50px;
  font-size: 200%;
}
div.mpld3-tooltip {
  background: #616161;
  color: white;
  font-size: 200%;
  opacity: 0.9;
  padding-left: 8dp;
  padding-right: 8dp;
}
"""

df3 = df2.sample(n=2000)
x=df3['PC2']
y=df3['PC2']

fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'))

scatter = ax.scatter(x, y, s=180, c="#90CAF9", )
plt.xlabel('PC2', size=20)
plt.ylabel('PC2', size=20)
ax.grid(color='white', linestyle='solid')
ax.set_title("Words by component vector", size=20)

labels = list(df3.index)
tooltip = mpld3.plugins.PointHTMLTooltip(scatter, labels=labels, 
                                         css=css)
mpld3.plugins.connect(fig, tooltip)
html = mpld3.fig_to_html(fig)

with open("lw_comp_23.html", "w") as f:
    f.write(html)
```
